In stock data, "Adj Close," "Close," "High," and "Low" are commonly used terms to describe different price points of a stock on a given day. Here's what each term represents:

Adj Close (Adjusted Close): The Adjusted Close price reflects the final price at which a stock trades for a particular day, taking into account any corporate actions or events that may have affected the stock's value. It incorporates adjustments such as stock splits, dividends, rights offerings, or spin-offs. The Adjusted Close is often considered the most accurate representation of a stock's value at the end of a trading day, as it accounts for these important factors.

Close: The Close price is the final price at which a stock is traded on a given trading day. It represents the last transaction price before the market closes for the day. The Close price is widely used to track the performance of a stock over time and is often used as a reference point for calculating returns, analyzing trends, and making investment decisions.

High: The High price represents the highest trading price that a stock reached during a particular trading day. It indicates the peak value that the stock attained before potentially pulling back or closing at a lower price. The High price is useful for identifying the intraday price range and gauging the stock's volatility.

Low: The Low price represents the lowest trading price that a stock reached during a specific trading day. It indicates the minimum value that the stock touched before potentially rebounding or closing at a higher price. The Low price is valuable for assessing the intraday price range and understanding the stock's volatility.

By analyzing these different price points, traders and investors can gain insights into a stock's performance, including its opening and closing prices, intraday fluctuations, and overall price movement over time.

```python
#Focusing on a single stock
#You can change the ticker
meta = close.META.copy().to_frame()
```

```python
meta
```

```python
meta.shift(periods=1)
```

```python
#In this juncture, I want to observe the change from the previous day to the next, so I shift by One period each
#Lag out Day (Lag1) represents values of the previous day 
meta['Lag1'] = meta.shift(periods=1)
```

```python
meta
```

```python
#So I need to subtract the data of Today(META) to that of Yesterday(Lag1)
meta.META.sub(meta.Lag1)
```

```python
#Today == "META" and Yesterday == "Lag1"
meta['Difference']= meta.META.sub(meta.Lag1)
```

```python
meta
```

```python
meta["% Difference"] = meta.META.div(meta.Lag1).sub(1).mul(100)
```

```python
meta
```

```python
meta.rename(columns= {'Lag1':"Yesterday"}, inplace=True)
meta.rename(columns= {'Difference':"Raw Daily Change"}, inplace=True)
meta.rename(columns= {'%Difference':"Daily Change(%)"}, inplace=True)
```

```python
meta
```

```python
# End of Business day per month 
# You can rename columns to reflect the changes
meta.resample("BM").last() 
```

```python
#Calculating the returns on a Daily Basis
#I need to delete the columns now
del meta["Yesterday"]
del meta["Raw Daily Change"]
del meta["% Difference"]
```

```python
#Removing all the NaN data and making it look more clean
ret = meta.pct_change().dropna()
```

When comparing a stock to SPY, we are essentially evaluating the performance of the stock in relation to the performance of the SPDR S&P 500 ETF (SPY). SPY is an exchange-traded fund that aims to track the performance of the S&P 500 index, which is a widely recognized benchmark for the overall U.S. stock market.

By comparing a stock with SPY, investors can assess how well the stock is performing relative to the broader market. Here are a few key aspects to consider when comparing a stock with SPY:

Returns: Comparing the total return of a stock with the return of SPY over the same period can provide insights into the stock's performance. Positive relative returns indicate that the stock outperformed the market, while negative relative returns suggest underperformance.

Volatility: Analyzing the volatility of a stock compared to SPY can help evaluate the stock's riskiness. If a stock exhibits higher volatility than SPY, it may be considered riskier, as it experiences larger price fluctuations.

Beta: Beta measures the sensitivity of a stock's returns to the returns of the broader market. A stock with a beta greater than 1 tends to be more volatile than the market, while a stock with a beta less than 1 is generally less volatile. By comparing a stock's beta with SPY, investors can gauge its volatility relative to the overall market.

Sector Performance: SPY represents a broad-based index, consisting of various sectors. Comparing the performance of a stock within its sector to the corresponding sector's performance in SPY can provide sector-specific insights.

Market Sentiment: Assessing how a stock moves in relation to SPY during market upswings and downturns can indicate its correlation with overall market sentiment. If a stock tends to rise or fall more than SPY during market movements, it might have higher or lower sensitivity to broader market trends.

It's important to note that comparing a stock to SPY is just one method of evaluating its performance. Investors may also consider other relevant factors such as fundamental analysis, company-specific news, industry trends, and individual investment goals when making investment decisions.

There are several other benchmarks used in financial markets to compare the performance of stocks or investment portfolios. Here are some commonly used benchmarks:

Dow Jones Industrial Average (DJIA): The DJIA is one of the oldest and most widely recognized stock market indices in the United States. It consists of 30 large, publicly traded companies that are considered industry leaders.

Nasdaq Composite Index: The Nasdaq Composite Index includes more than 3,000 stocks listed on the Nasdaq Stock Market. It primarily represents technology companies but also includes stocks from various other sectors.

Russell 2000 Index: The Russell 2000 Index measures the performance of approximately 2,000 small-cap stocks in the United States. It provides insight into the performance of smaller companies, which can differ from larger-cap stocks represented by indices like the S&P 500.

MSCI World Index: The MSCI World Index is a global equity index that tracks the performance of stocks from developed countries around the world. It is often used as a benchmark for global equity investments.

FTSE 100 Index: The FTSE 100 Index is a benchmark index of the 100 largest companies listed on the London Stock Exchange. It represents a broad range of industries in the United Kingdom.

Nikkei 225 Index: The Nikkei 225 is a stock market index that tracks the performance of 225 large, publicly traded companies listed on the Tokyo Stock Exchange. It is one of the most widely followed benchmarks in Japan.

Hang Seng Index: The Hang Seng Index is a market capitalization-weighted index that tracks the performance of the largest and most liquid companies listed on the Hong Kong Stock Exchange.

These are just a few examples of benchmarks used in different regions and for specific market segments. It's important to choose a benchmark that aligns with the specific investment objectives and geographic focus of the stocks or portfolios being evaluated.

```python
return_all = close_all.pct_change().dropna()
```

```python
return_all.describe().T
```

```python
summary_close_all = return_all.describe().T.loc[:, ["mean", "std"]]
```

```python
summary_close_all
```

```python
# Annual summaries for the percentage change..i.e what should I expect as returns per year
summary_close_all["mean"] =  summary_close_all["mean"] * 252
summary_close_all["std"] =  summary_close_all["std"] * np.sqrt(252)
summary_close_all
```

```python
import plotly.express as px

fig = px.scatter(summary_close_all, x='std', y='mean', title='Scatter Plot')
fig.update_layout(width=800, height=600, xaxis_title='Annual Risk (Standard Deviation)', yaxis_title='Annual Returns', title = "Risk / Return Graph")

for i in summary_close_all.index:
    fig.add_annotation(
        x=summary_close_all.loc[i, "std"] + 0.02,
        y=summary_close_all.loc[i, "mean"] + 0.02,
        text=str(i),
        showarrow=False,
        font=dict(size=15),
    )

fig.show()
#We can see below how NVDA has higher returns but also high risk
#For short term, we can Invest in NVDA
#For Long term , we can invest in MSFT, GOOG and AAPL
#We may Avoid Meta & AMZN now
```
